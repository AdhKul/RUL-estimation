```python
def find_rms(df1):
    sum=0
    df2=df1.to_numpy()
    for i in df2:
        sum=sum+pow((i),2)
    sum=sum/(len(df1.index))
    return(pow(sum,0.5))
        
```

```python
def statistics_b(df1): 
    bmean=df1['B'].mean()
    bmax=df1['B'].max()
    bmin=df1['B'].min()
    bmedian=df1['B'].median()
    bmode1=df1['B'].mode()
    if(bmode1[0]):
        bmode=bmode1[0]
    else:                                #multiple modes can exist so taking the first value
        bmode=float(df1['B'].mode())    
    bstddev=df1['B'].std()
    bsamvar=df1['B'].var()
    bkurt=df1['B'].kurt()
    bskew=df1['B'].skew()
    brange=df1['B'].max() - df1['B'].min()
    bsum=df1['B'].sum()
    baxisid=1 #corresponds to the 'B' column 
    bcondition=0 #needs to be updated based on the file from which the data is extracted
    brms=find_rms(df1['B']) #RMS doubt
    blist=[]
    blist.extend((bmean, bmax, bmin, bmedian, bmode, bstddev, bsamvar, bkurt, bskew, brange, bsum, brms, baxisid))
    return blist
```

```python
def statistics_d(df1): 
    dmean=df1['D'].mean()
    dmax=df1['D'].max()
    dmin=df1['D'].min()
    dmedian=df1['D'].median()
    dmode1=df1['D'].mode()
    if(dmode1[0]):
        dmode=dmode1[0]
    else:                                #multiple modes can exist so taking the first value
        dmode=float(df1['D'].mode())
    dstddev=df1['D'].std()
    dsamvar=df1['D'].var()
    dkurt=df1['D'].kurt()
    dskew=df1['D'].skew()
    drange=df1['D'].max() - df1['D'].min()
    dsum=df1['D'].sum()
    daxisid=2 #corresponds to the 'D' column 
    dcondition=0 #needs to be updated based on the file from which the data is extracted
    drms=find_rms(df1['D']) #RMS doubt
    dlist=[]
    dlist.extend((dmean, dmax, dmin, dmedian, dmode, dstddev, dsamvar, dkurt, dskew, drange, dsum, drms, daxisid))
    return dlist
```

```python
def statistics_f(df1): 
    fmean=df1['F'].mean()
    fmax=df1['F'].max()
    fmin=df1['F'].min()
    fmedian=df1['F'].median()
    fmode1=df1['F'].mode()
    if(fmode1[0]):
        fmode=fmode1[0]
    else:                                #multiple modes can exist so taking the first value
        fmode=float(df1['F'].mode())
    fstddev=df1['F'].std()
    fsamvar=df1['F'].var()
    fkurt=df1['F'].kurt()
    fskew=df1['F'].skew()
    frange=df1['F'].max() - df1['F'].min()
    fsum=df1['F'].sum()
    faxisid=3 #corresponds to the 'F' column 
    fcondition=0 #needs to be updated based on the file from which the data is extracted
    frms=find_rms(df1['F']) #RMS doubt
    flist=[]
    flist.extend((fmean, fmax, fmin, fmedian, fmode, fstddev, fsamvar, fkurt, fskew, frange, fsum, frms, faxisid))
    return flist
```

```python
def statistics_h(df1): 
    hmean=df1['H'].mean()
    hmax=df1['H'].max()
    hmin=df1['H'].min()
    hmedian=df1['H'].median()
    hmode1=df1['H'].mode()
    if(hmode1[0]):
        hmode=hmode1[0]
    else:                                #multiple modes can exist so taking the first value
        hmode=float(df1['H'].mode())
    hstddev=df1['H'].std()
    hsamvar=df1['H'].var()
    hkurt=df1['H'].kurt()
    hskew=df1['H'].skew()
    hrange=df1['H'].max() - df1['H'].min()
    hsum=df1['H'].sum()
    haxisid=4 #corresponds to the 'H' column 
    hcondition=0 #needs to be updated based on the file from which the data is extracted
    hrms=find_rms(df1['H']) #RMS doubt
    hlist=[]
    hlist.extend((hmean, hmax, hmin, hmedian, hmode, hstddev, hsamvar, hkurt, hskew, hrange, hsum, hrms, haxisid))
    return hlist
```

```python
def get_statistics_for_a_file(file_name1,file_number):
    df = pd.read_csv (file_name1,header = None, names = columns)
    df=df.drop(['A','C','E','G'], axis=1) #dropping time-series columns
    rows_count= len(df.index)
    print(rows_count)
    upper_limit=int(rows_count/96000)
    print(upper_limit) #all the excel files have different number of rows
    condition=file_number-745
    for i in range(0,upper_limit):
        df_trim=df.iloc[(96000*i):96000*(i+1)]
        ls_b=statistics_b(df_trim)
        ls_b.append(condition)
        data.loc[len(data)] = ls_b
        ls_d=statistics_d(df_trim)
        ls_d.append(condition)
        data.loc[len(data)] = ls_d
        ls_f=statistics_f(df_trim)
        ls_f.append(condition)
        data.loc[len(data)] = ls_f
        ls_h=statistics_h(df_trim)
        ls_h.append(condition)
        data.loc[len(data)] = ls_h
    
    df_trim=df.iloc[(96000*upper_limit):] #the no of rows is not an exact multiple of 96000
    ls_b=statistics_b(df_trim)
    ls_b.append(condition)
    data.loc[len(data)] = ls_b
    ls_d=statistics_d(df_trim)
    ls_d.append(condition)
    data.loc[len(data)] = ls_d
    ls_f=statistics_f(df_trim)
    ls_f.append(condition)
    data.loc[len(data)] = ls_f
    ls_h=statistics_h(df_trim)
    ls_h.append(condition)
    data.loc[len(data)] = ls_h   
```

# Similarity Matching


```python
def euclidean_distance( a,b):
    sum=0
    for i in range (0,L):
        sum=sum+(pow((a[i]-b[i]),2))
    return(pow(sum,0.5))
```

```python
def cosine_distance(a,b):
    sum1=0
    sum2=0
    numerator=0
    for i in range (0,L):
        numerator=numerator+(a[i]*b[i])
        sum1=sum1+(pow(a[i],2))
        sum2=sum2+(pow(b[i],2))
    sum3=pow(sum1,0.5)
    sum4=pow(sum2,0.5)
    denominator=sum3*sum4
    fraction=numerator/denominator
    return((1-fraction))
```

```python
def inverse(a):
    return(pow(a,-1))
```

```python
y_pred=np.array([])
y_pred
```

```python
no_of_segments=8060-L+1
distance_array=np.array([])
x_te=X_test[16:2016]
for i in range(0,no_of_segments): 
    lower_limit=i;
    upper_limit=(i+L)
    x_sel=X_train[lower_limit:upper_limit]
    print(i)
    print("x_sel:",x_sel.shape)
    print("x_te:",x_te.shape)
    ed=euclidean_distance(x_te,x_sel)
    cd=cosine_distance(x_te,x_sel)
    distance=inverse(ed)+inverse(cd)
    distance_array=np.append(distance_array,distance)

min_index=np.argmin(distance_array,axis=0)
y_pr=y_train_arr[min_index+L-1]
y_pred=np.append(y_pred,y_pr)
```
